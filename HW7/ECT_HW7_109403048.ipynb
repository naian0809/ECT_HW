```python
import numpy as np
import pandas as pd
```

第一題

```python
df = pd.read_csv("economic_freedom_data.csv")
df
```

```python
df.info() #可以看到全部都沒有遺漏值
```

```python
df.describe().T #每一個欄位的平均值和標準差都有一定差距，表明都要列為關係欄位
```

```python
df['freedom'].unique() #可以看到target是object 並只有兩個值，freedom 和 not-freedom
```

```python
df['freedom'] = df['freedom'].replace('not-freedom', 0).replace('freedom', 1) #替換freedom為1 not-freedom為0
```

```python
df_shuffled = df.sample(frac=1).reset_index(drop=True) #將排序洗牌
```

```python
y = df_shuffled['freedom']
X = df_shuffled.drop(['freedom'], axis=1) #分為target和關係欄位
```

第二題

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import f1_score
```

```python
# RandomForest
rfc = RandomForestClassifier(random_state=10)
rfc_f1_scores = cross_val_score(rfc, X, y, cv=10)
average_rfc_f1_score = rfc_f1_scores.mean()
print("RandomForest Average F1 Score:", average_rfc_f1_score)
```

第三題

```python
from sklearn.neural_network import MLPClassifier
mlp_classifier = MLPClassifier(hidden_layer_sizes=(64, 128), random_state=10, batch_size=8)
mlp_f1_scores = cross_val_score(mlp_classifier, X, y, cv=10, scoring='f1_macro')
average_mlp_f1_score = mlp_f1_scores.mean()
print("Scikit-learn MLP Average F1 Score:", average_mlp_f1_score)
```

第四題

```python
import tensorflow
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import StratifiedKFold
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(64, input_dim=X.shape[1], activation='relu'))
    model.add(Dense(128, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model


kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=10)
k_mlp_f1_scores = []
for train_index, test_index in kfold.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]


    model = create_model()


    model.fit(X_train, y_train, epochs=10, batch_size=8, verbose=0)

    y_pred_prob = model.predict(X_test)
    y_pred = np.where(y_pred_prob > 0.5, 1, 0)

    f1 = f1_score(y_test, y_pred)


    k_mlp_f1_scores.append(f1)


# 打印平均 F1 分数
average_k_mlp_f1_score = np.mean(k_mlp_f1_scores)
print("Keras MLP Average F1 Score:", average_k_mlp_f1_score)
```

第六題

```python
from scipy.stats import ttest_ind



t_statistic, p_value = ttest_ind(rfc_f1_scores,mlp_f1_scores )
t_statistic2, p_value2 = ttest_ind(rfc_f1_scores, k_mlp_f1_scores)
t_statistic3, p_value3 = ttest_ind(mlp_f1_scores,k_mlp_f1_scores )
 
print("Comparison between RandomForest and scikit-learn MLP:")
print("t-statistic:", t_statistic)
print("p-value:", p_value)
print()
print("Comparison between RandomForest and Keras MLP:")
print("t-statistic:", t_statistic2)
print("p-value:", p_value2)
print()
print("Comparison between scikit-learn MLP and Keras MLP:")
print("t-statistic:", t_statistic3)
print("p-value:", p_value3)
```

以95%的信心水準來看
隨機森林（RandomForest）和 scikit-learn 的多層感知器（MLP）之間的比較：
t 統計值為 1.993835331813589，p 值為 0.06154624709235281。p>0.05
根據統計結果，我們無法拒絕兩者在表現上沒有顯著差異的假設。

隨機森林（RandomForest）和 Keras 的多層感知器（MLP）之間的比較：
t 統計值為 4.617896154977763，p 值為 0.00021367632798739768。
根據統計結果，我們可以拒絕兩者在表現上沒有顯著差異的假設。這意味著隨機森林和 Keras 的 MLP 在某些方面上可能有顯著的差異。

scikit-learn 的多層感知器（MLP）和 Keras 的多層感知器（MLP）之間的比較：

根據統計結果，我們可以拒絕兩者在表現上沒有顯著差異的假設。這意味著 scikit-learn 的 MLP 和 Keras 的 MLP 在某些方面上可能有顯著的差異。
