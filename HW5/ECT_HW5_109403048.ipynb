# 第六題

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.exceptions import ConvergenceWarning
import warnings

class ChurnPrediction:
    
    def __init__(self):
        self.stratified_sampling_df = None
        self.le = LabelEncoder()
        self.scaler = StandardScaler()
        self.log_reg_model = LogisticRegression(max_iter=2000)
        self.svm_model = SVC(max_iter=2000)
        
    def run_pipeline(self):
        self.load_data()
        self.stratified_sampling()
        self.fill_null()
        self.remove_col()
        self.one_hot_encoding()
        self.label_encoding()
        self.evaluate_models()
        
    def load_data(self):
        raw_data = pd.read_csv('customer_churn.csv')
        self.stratified_sampling_df = pd.DataFrame(raw_data)
        
    def stratified_sampling(self):
        self.stratified_sampling_df = self.stratified_sampling_df.groupby('Churn', group_keys=False).apply(lambda x: x.sample(frac=0.6))

    def fill_null(self):
        mean_values = self.stratified_sampling_df.mean()
        self.stratified_sampling_df.fillna(value=mean_values, inplace=True)
        
    def remove_col(self):
        self.stratified_sampling_df = self.stratified_sampling_df.drop('CustomerID', axis=1)
        
    def one_hot_encoding(self):
        self.stratified_sampling_df = pd.get_dummies(self.stratified_sampling_df, columns=['PreferredLoginDevice', 'PreferredPaymentMode', 'Gender', 'PreferedOrderCat', 'MaritalStatus'])
        
    def label_encoding(self):
        self.stratified_sampling_df['Churn'] = self.le.fit_transform(self.stratified_sampling_df['Churn'])
        
    def evaluate_models(self):
        X = self.stratified_sampling_df.drop(['Churn'], axis=1)
        y = self.stratified_sampling_df['Churn']

        X_scaled = self.scaler.fit_transform(X)

        with warnings.catch_warnings():
            warnings.simplefilter("ignore", category=ConvergenceWarning)

            log_reg_scores = cross_val_score(self.log_reg_model, X_scaled, y, cv=10)
            svm_scores = cross_val_score(self.svm_model, X_scaled, y, cv=10)

        self.log_reg_mean_score = np.mean(log_reg_scores)
        self.svm_mean_score = np.mean(svm_scores)
```

```python
num_runs = 30
log_reg_scores = []
svm_scores = []

for i in range(num_runs):
    churn_prediction = ChurnPrediction()
    churn_prediction.run_pipeline()
    
    log_reg_scores.append(churn_prediction.log_reg_mean_score)
    svm_scores.append(churn_prediction.svm_mean_score)
    
print(f"Logistic Regression Mean Score: {np.mean(log_reg_scores)}")
print(f"SVM Mean Score: {np.mean(svm_scores)}")
```

# 第七題

```python
from scipy import stats

# Calculate mean scores
log_reg_mean = np.mean(log_reg_scores)
svm_mean = np.mean(svm_scores)

# Calculate standard deviations
log_reg_std = np.std(log_reg_scores)
svm_std = np.std(svm_scores)

# Calculate t-value and p-value
t, p = stats.ttest_rel(log_reg_scores, svm_scores)

# Print results
print('Logistic Regression mean:', log_reg_mean, 'std:', log_reg_std)
print('SVM mean:', svm_mean, 'std:', svm_std)
print('t-value:', t)
print('p-value:', p)
```

p-value的值為4.111236931855208e-18，小於0.05的顯著性水平，表示可以拒絕虛無假設(兩種模型的平均準確率是相等的)，並且接受對立假設(兩種模型的平均準確率是不相等的)。因此，由logistic regression 與SVM的輸出結果我們可以得出結論：SVM模型在此分類任務中表現比Logistic Regression更好。
