第五題

```python
df_copy = df.copy()

df_copy.loc[df['freedom_own_foreign_currency'] == 'HIGH', 'freedom_own_foreign_currency'] = 2
df_copy.loc[df['freedom_own_foreign_currency'] == 'NORMAL', 'freedom_own_foreign_currency'] = 1
df_copy.loc[df['freedom_own_foreign_currency'] == 'LOW', 'freedom_own_foreign_currency'] = 0
df.update(df_copy)
```

第六題

```python
df_copy = df.copy()
df_copy.loc[df_copy['eco_freedom']<6,'eco_freedom']=0
df_copy.loc[df_copy['eco_freedom']>=6,'eco_freedom']=1
df.update(df_copy)
```

```python
df['eco_freedom']=df['eco_freedom'].astype(int)#原先是float改成int
```

```python
df.rename(columns={'eco_freedom': 'freedom'}, inplace=True)#改名freedom
```

```python
df['freedom'].value_counts()#檢查有無除了1、0以外的值
```

第七題

```python
mean_values = df.mean()
df.fillna(value=mean_values, inplace=True)#塞入該欄的平均
```

```python
df
```

```python
null_counts = df.isnull().sum()#檢查還有沒有空值
print(null_counts)
```

第八題

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from sklearn import tree
```

```python
new_df = df.loc[:, ~df.columns.isin(['year', 'iso_code','countries','rank','freedom'])]
y = df['freedom']
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_df, y, train_size=0.5, random_state=12,stratify=y)

#創建決策樹模型
dt= DecisionTreeClassifier(criterion='gini', max_leaf_nodes=5)
#訓練
dt.fit(X_train, y_train)
# 預測
y_pred = dt.predict(X_test)
```

```python
plt.figure(figsize=(10,10))
_=tree.plot_tree(dt,filled = True, feature_names=X_train.columns)
plt.show()
```

第九題

```python
accuracy = accuracy_score(y_test, y_pred)

print("平均準確率：", accuracy)
```

第十題

```python
target_names = ['non-freedom','freedom']
print(classification_report(y_test, y_pred,target_names = target_names ))
```

第十一題

```python
#size_ratio為 calculate_acuracy() 的input 帶入ratio_array return該比例的準確率
def calculate_acuracy(size_ratio):
    X_train, X_test, y_train, y_test = train_test_split(new_df, y, train_size=size_ratio, random_state=12, stratify=y)
    dt= DecisionTreeClassifier(max_leaf_nodes=5)
    dt.fit(X_train, y_train)
    y_pred = dt.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy
ratio_array = [0.5, 0.6, 0.7, 0.8, 0.9]
score=[]
for i in ratio_array:
       score=np.append(score,calculate_acuracy(i))
```

```python
spilit_proportion = ['50/50','60/40','70/30','80/20','90/10']
acuracy_df = pd.DataFrame({'split_proportion':spilit_proportion , 'score': score})
acuracy_df
```

第十二題

```python
import matplotlib.pyplot as plt

x_data = acuracy_df['split_proportion'].values
y_data = acuracy_df['score'].values

plt.plot(x_data, y_data,color='red', label='score')

plt.title('train/test split proportion & result',fontsize='12',weight='semibold',pad=20)
plt.legend()
plt.show()
```
